```python
dataset=dataset.drop(columns=['months_employed'])
```

```python
dataset['personal_account_months']=(dataset.personal_account_m+ (dataset.personal_account_y*12))
```

```python
dataset[['personal_account_m','personal_account_y','personal_account_months']].head()
```

```python
dataset=dataset.drop(columns=['personal_account_m','personal_account_y'])
```

```python
dataset.head()
```

# Onehotencoding

```python
dataset=pd.get_dummies(dataset)
```

```python
dataset.head()
```

```python
dataset=dataset.drop(columns=['pay_schedule_semi-monthly'])
```

# Remove extra columns


```python
y=dataset["e_signed"]
id1=dataset["entry_id"]
dataset=dataset.drop(columns=['e_signed','entry_id'])
```

```python
dataset.head()
```

# splitting data to train and test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(dataset, y, test_size=0.2, random_state=2)
```

# feature scaling


```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_X=StandardScaler()
```

```python
X_train2=pd.DataFrame(sc_X.fit_transform(X_train))
X_test2=pd.DataFrame(sc_X.fit_transform(X_test))
```

```python
X_test2
```

```python
X_train2.columns=X_train.columns.values
X_test2.columns=X_test.columns.values
```

```python
X_train2.index=X_train.index.values
X_test2.index=X_test.index.values
```

```python
X_train=X_train2
X_test=X_test2
```

```python
X_train
```

# Logistic regression


```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,f1_score,precision_score,recall_score
```

```python
acc=accuracy_score(y_test,y_pred)
prec=precision_score(y_test,y_pred)
rec=recall_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)
```

```python
results=pd.DataFrame([["Logistic Regression",acc,prec,rec,f1]],columns=['Model','Accuracy','Precision','Recall','f1_score'])
```

# SVM classifier

```python
from sklearn.svm import SVC
classifier=SVC(random_state=1,kernel='rbf')
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
acc=accuracy_score(y_test,y_pred)
prec=precision_score(y_test,y_pred)
rec=recall_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)
```

```python
svc_results=pd.DataFrame([["SVC Classifier_rbf",acc,prec,rec,f1]],columns=['Model','Accuracy','Precision','Recall','f1_score'])
```

```python
svc_results
```

```python
results=results.append(svc_results,ignore_index=True)
```

```python
results
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
rfc=RandomForestClassifier()
param_grid={
    'n_estimators':[200,700],
    'max_features':['auto','sqrt','log2']
}
cv_rfc=GridSearchCV(estimator=rfc,param_grid=param_grid,cv=5)
cv_rfc.fit(X_train,y_train)
```

```python
y_pred=cv_rfc.predict(X_test)
```

```python
acc=accuracy_score(y_test,y_pred)
prec=precision_score(y_test,y_pred)
rec=recall_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)
```

```python
cv_results=pd.DataFrame([["rfc Classifier",acc,prec,rec,f1]],columns=['Model','Accuracy','Precision','Recall','f1_score'])
```

```python
results=results.append(cv_results,ignore_index=True)
```
